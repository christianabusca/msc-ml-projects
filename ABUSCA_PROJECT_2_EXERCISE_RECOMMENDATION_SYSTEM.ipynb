**Import libraries**

```python
#import pandas and sklearn
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

```

**Read file**

```python
# Use pd.read_excel to read excel file
df = pd.read_excel('sample_data/gym-recommendation.xlsx')
df.head()

# Gender = male = 1 female = 0
# Fitness Type = Muscular fitness = 0 Cardiovascular fitness = 1
```

**Seperate X and y columns**

```python
# Create variables each for X and y , then subset the gym data
# Input must have the gender, bmi, and fitness type columns, and output is the exercise column
x = df[['Gender', 'BMI', 'Fitness Type']]
y = df['Exercises']

x.head()
y.head()
```

**Split Data**

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

#Set random state to 0 and the test size to 0.2
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=0)
print(f"Training set shape: {x_train.shape}, {y_train.shape}")
print(f"Test set shape: {x_test.shape}, {y_test.shape}")
```

**Train the model**

```python
# Create a variable named model and choose an algorithm to train it, the fit the training data
model = LinearRegression()
model.fit(x_train, y_train)
```

**Test the model**

```python
# Create a variable named ytestpredictions and then predict the model using the X_test data
ytestpredictions = model.predict(x_test)
ytestpredictions_rounded = [round(pred) for pred in ytestpredictions]

```

**Compare actual and predicted values**

```python
# Create a variable named predictions and create a DataFrame of the actual and predictions data using y_test and ytestpredictions
# join the X_test with the predictions DataFrame and show the created DataFrame
predictions = pd.DataFrame({'Actual': y_test, 'Predicted': ytestpredictions_rounded})
predictions = pd.concat([x_test.reset_index(drop=True), predictions.reset_index(drop=True)], axis=1)
print(predictions)
```

**Check accuracy score**

```python
# Create a variable named score and then use y_test and ytestpredictions to check the accuracy score, then show the score
score = accuracy_score(y_test, ytestpredictions_rounded)
print(f"Accuracy Score: {score}")
```

**Use the model**

```python
# Try predicting using unseen data in the order of: gender, bmi, fitness type
# example : model.predict ([[0, 19, 1], [1, 27, 0]])
model.predict ([[0, 19, 1], [1, 27, 0]])
```
